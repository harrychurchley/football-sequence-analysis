```python
from statsbombpy import sb
```

```python
# view all competitions and seasons. Free data only
sb.competitions()
```

```python
# select desired data
sb.matches(competition_id=49, season_id=3)
```

```python
import warnings
warnings.filterwarnings('ignore')

events = sb.competition_events(
    country="England",
    division= "FA Women's Super League",
    season="2020/2021",
    gender="female"
)
```

```python
import pickle

with open('wsl_20-21.pkl', 'wb') as f:
    pickle.dump(events, f)
```

```python
import bz2

with open('wsl_20-21_compressed.pkl', 'wb') as f:
    compressed_file = bz2.BZ2File(f, 'w')
    pickle.dump(events, compressed_file)
```
